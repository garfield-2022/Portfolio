# Data Analysis

```python
# import necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")
%matplotlib inline
```

```python
# load the dataset into a pandas dataframe
data_path = 'creditcard.csv'
df = pd.read_csv(data_path)
```

```python
# The dataset consists of 31 columns. v1-V28 are anonymous and transformed. Time, Amount and Class are untouched.
# The goal of data analysis is to explore the relationship between Class and the other 30 colomns.
df.head()
```

```python
df.info()
```

```python
# checking the number of missing values in each column
# no missing values in the dataset
df.isnull().sum()
```

```python
# Statistical summary of each colomn.
# There are 284807 rows in the dataset.
df.describe()
```

```python
# Discretization and binning.
bins = [0, 1000, 2000, 3000, 4000, 5000, 26000]
cats = pd.cut(df.Amount,bins)
cats
```

```python
# The distribution of Amount column. Most of them are within (0,1000]. Only 55 are more than 5000.
pd.value_counts(cats)
```

```python
sns.histplot(df.Amount, bins=50, color='b')
```

```python
# check the distribution of the class column
# 0: normal, 1: anomaly
# the data is overwhelmingly biased towards normal data
sns.histplot(data=df, y="Class", bins=50, color='b')
```

```python
# separate the dataset into two
anomalies = df[df.Class == 1]
normal = df[df.Class == 0]
print(f"Anomalies: {anomalies.shape}")
print(f"Normal: {normal.shape}")
```

```python
# statistical features of Amount attribute in the anomaly sub-dataset
anomalies.Amount.describe()
```

```python
# statistical features of Amount attribute in the normal sub-dataset
normal.Amount.describe()
```

```python
def get_stats(group):
    return {'min': group.min(), 'max': group.max(), 'count':group.count(), 'mean': group.mean()}

df.groupby(by=['Class']).apply(get_stats)
```

```python
# Normal transaction amount lies in a very large range, while fraud much smaller.
sns.relplot(data=df, x="Amount", y="Class")
```

```python
# Both normal and anomaly transactions happen across all the time span.
fig, axes = plt.subplots(1, 2, sharex=True, sharey=True)
axes[0].set_title('Normal')
sns.scatterplot(ax=axes[0], data=normal, x="Time", y="Amount", color='b')

axes[1].set_title('Anomaly')
sns.scatterplot(ax=axes[1], data=anomalies, x="Time", y="Amount", color='r')
```

```python
# Over the timespan, the normal transations show some pattern.
# Compared with normal transactions above, the anomalies' pattern is not very clear, 
# although it overlaps with that of normal at around 100000.
fig, axes = plt.subplots(2, 1, sharex=True)
axes[0].set_title('Normal')
sns.histplot(ax=axes[0], data=normal, x="Time", color='b')

axes[1].set_title('Anomaly')
sns.histplot(ax=axes[1], data=anomalies, x="Time", color='r')
```

```python
# now look at how all of these data points vary according to time:
# Although a significant portion of the anomalies fall under the band of normal points, 
# there are still a good number of anomalies that fall out of that range. 
# Against Time, the data for the column V12 also shows this deviation from the normal data points
# For V17, the difference seems to be a bit more pronounced, as the anomalies seem to be more spread out. 
for f in range(1, 29):
    fig, axes = plt.subplots(1, 2, sharex=True, sharey=True)
    axes[0].set_title('Normal')
    sns.scatterplot(ax=axes[0], data=normal, x="Time", y=f'V{f}', color='b')

    axes[1].set_title('Anomaly')
    sns.scatterplot(ax=axes[1], data=anomalies, x="Time", y=f'V{f}', color='r')
```
